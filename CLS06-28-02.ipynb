```python
ser1 = pd.Series([1, 2, 3, 4, 5])
ser2 = pd.Series([3, 4, 5, 6, 7])
print(ser1)
print(ser2)
```

```python
union = pd.Series(np.union1d(ser1, ser2))
intersect = pd.Series(np.intersect1d(ser1, ser2))
notcommonseries = union[~union.isin(intersect)]
```

```python
plt.figure(figsize=(12,12))
plt.plot(x,y,color = "black",marker ='D',linewidth = 2, label='sinx')
plt.plot(x,np.sin(x),color = "red",marker ='o',linewidth = 2, label='cosx')
plt.xlabel('values')
plt.ylabel('sinx')
plt.title('graphs')
#plt.grid(True)
plt.legend(loc = 'upper right' )
for i,j in zip(x,y):
    plt.text (i,j, str(round(j,2)), color='black', fontsize= 12)
    
for k,h in zip(x,np.sin(x)):
    plt.text (k,h, str(round(h,2)),color='red',fontsize=12)
    
```
